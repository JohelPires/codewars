```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python

# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv(
    '/home/johel/Documentos/codewars_fns/codewars/page-view-time-series-visualizer/fcc-forum-pageviews.csv', index_col=0, parse_dates=True)

df

```

```python
q = df.value.quantile(0.025)
upperq = df.value.quantile(0.975)
df = df[(df.value > q) & (df.value < upperq)]
```

## Draw Line Plot

```python
# df.plot()

# cleandf.plot(fmt='r', title='Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
plt.figure(figsize=(15, 5))
plt.plot(df, 'r', linewidth=1)

plt.title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
plt.xlabel("Dates")
plt.ylabel("Page Views")

plt.show()
```

## Draw Bar Plot

```python
df.groupby([df.index.year, df.index.month]).value.mean().plot.bar(legend=True)
plt.legend(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])
```
