```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/johel/Documentos/codewars_fns/codewars/Piece_Dim.csv', index_col='Item_No')
```

```python
sns.set_style('whitegrid')
```

```python
sns.histplot(data=df, x='Length', bins=10)
plt.title('Length Histogram')
plt.show()
```

```python
sns.displot(data=df, x='Length', bins=10, hue='Operator')
plt.title('Length Histogram')
plt.show()
```

```python
sns.displot(data=df, x='Length', bins=10, col='Operator')
# plt.title('Length Histogram')
plt.show()
```

```python
sns.displot(data=df, x='Length', rug=True, kind='kde', hue='Operator')
plt.title('Length kde')
plt.show()
```

```python
sns.displot(data=df, x='Length', kind='ecdf', rug=True, hue='Operator')
plt.title('Length ecdf')
plt.show()
```

```python
sns.displot(data=df, x='Length', kind='hist', rug=True, kde=True, col='Operator')
# plt.title(['Length Histogram', 'op2'])
plt.show()
```

```python
sns.displot(data=df, x='Length', y='Width', kind='kde')
```
